```python
import zipfile
import pandas as pd
import os  # Add this line to import the os module

zip_file_path = '/Users/eduardoaleman/Desktop/Assignments/citibike/202301-citibike-tripdata.csv.zip'
csv_file_name = '202301-citibike-tripdata.csv'
extracted_folder = '/Users/eduardoaleman/Desktop/Assignments/citibike/extracted_files/'

os.makedirs(extracted_folder, exist_ok=True)

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extract(csv_file_name, extracted_folder)

csv_file_path = os.path.join(extracted_folder, csv_file_name)
NOW = pd.read_csv(csv_file_path, low_memory=False)

NOW.head()
```

```python
replace_dict = {'member': 'Subscriber', 'casual': 'Customer'}
NOW['member_casual'] = NOW['member_casual'].replace(replace_dict)
```

```python
NOW.head()
```

```python
replace_dict = {'classic_bike': 'Classic', 'electric_bike': 'Electric', 'docked_bike': 'Docked'}
NOW['rideable_type'] = NOW['rideable_type'].replace(replace_dict)
```

```python
NOW.head()
```

```python
NOW[['2023_Date', '2023_Time']] = NOW['started_at'].str.split(' ', 1, expand=True)
NOW[['2023_End_Date', '2023_End_Time']] = NOW['started_at'].str.split(' ', 1, expand=True)
```

```python
NOW.head()
```

```python
NOW.head()
```

```python
NOW.rename(columns={'start_station_name': 'start station name',
                    'end_station_name': 'end station name',
                    'start_station_id': 'start station id',
                    'end_station_id': 'end station id',
                    'start_lat': 'start station latitude',
                    'start_lng': 'start station longitude',
                    'end_lat': 'end station latitude',
                    'end_lng': 'end station longitude',
                    'member_casual': 'usertype',
                    '2023_End_Date': 'Date2',
                    '2023_End_Time': 'End_Time'}, inplace=True)

```

```python
NOW.head()
```

```python
NOW.rename(columns={'2023_Date': 'Date'}, inplace=True)
```

```python
NOW.rename(columns={'2023_Time': 'Start_Time'}, inplace=True)
```

```python
NOW.head()
```

```python
NOW = NOW.assign(tripduration=pd.NA)
```

```python
NOW = NOW.assign(Age=pd.NA)
```

```python
NOW = NOW.assign(Gender=pd.NA)
```

```python
NOW.head()
```

```python
NOW = NOW.drop(['ride_id'], axis=1)
```

```python
NOW.head()
```

```python
csv_file_path = '2023.csv'
NOW.to_csv(csv_file_path, index=False)
```
